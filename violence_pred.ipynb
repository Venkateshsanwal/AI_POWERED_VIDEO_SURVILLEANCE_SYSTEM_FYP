```python
import numpy as np
import argparse
import pickle
import cv2
import os
import time 
from keras.models import load_model
from collections import deque

def print_results(video, limit=None):
    if not os.path.exists('output'):
        os.mkdir('output')

    print("Loading model ...")
    model = load_model('/Users/venkateshsanwal/Desktop/fyp/violence_detection/real_life_vd/archive/violence_detection_model.h5')
    
    # Initialize deque for prediction history
    Q = deque(maxlen=128)

    # Open the video file
    vs = cv2.VideoCapture(video)

    # Check if the video file was opened correctly
    if not vs.isOpened():
        print(f"Error: Could not open video file {video}")
        return

    writer = None
    (W, H) = (None, None)
    count = 0     
    while True:
        # Read the next frame from the file
        (grabbed, frame) = vs.read()

        # If the frame was not grabbed, we have reached the end
        if not grabbed:
            break

        # If the frame dimensions are empty, grab them
        if W is None or H is None:
            (H, W) = frame.shape[:2]

        # Clone the output frame, then convert it from BGR to RGB, resize, and perform mean subtraction
        output = frame.copy()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (128, 128)).astype("float32")
        frame = frame.reshape(128, 128, 3) / 255.0

        # Make predictions on the frame and update the predictions queue
        preds = model.predict(np.expand_dims(frame, axis=0))[0]
        Q.append(preds)

        # Perform prediction averaging over the history of previous predictions
        results = np.array(Q).mean(axis=0)
        i = (preds > 0.50)[0]  # If the prediction is > 50% for Violence, set label to 1 (Violence)
        label = i

        # Set text color based on the label
        text_color = (0, 255, 0)  # Default: green for non-violence
        if label:  # Violence predicted
            text_color = (0, 0, 255)  # Red for violence

        text = "Violence: {}".format(label)
        FONT = cv2.FONT_HERSHEY_SIMPLEX

        # Draw text on the output frame
        cv2.putText(output, text, (35, 50), FONT, 1.25, text_color, 3)

        # Check if the video writer is None (first frame)
        if writer is None:
            # Initialize video writer only if the video capture is successful
            fourcc = cv2.VideoWriter_fourcc(*"MJPG")
            writer = cv2.VideoWriter("output/v_output.avi", fourcc, 30, (W, H), True)

        # Write the output frame to disk (only after writer is initialized)
        if writer is not None:
            writer.write(output)

        # Show the output frame (replace cv2_imshow with cv2.imshow for local environment)
        cv2.imshow("Output", output)
        
        
        # Wait for 1ms, exit on 'q' key press
        key = cv2.waitKey(1) & 0xFF
        if key == ord("q"):
            break

    # Release resources after processing is complete
    print("[INFO] cleaning up...")
    if writer is not None:
        writer.release()
    vs.release()
    cv2.destroyAllWindows()  # Close all OpenCV windows
```
